# books

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('book (1).csv')
data
```

```python
items=apriori(data,min_support=0.1,use_colnames=True)
items
```

```python
rules=association_rules(items,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(data=rules,x='support',y='confidence')
```

# my_movies

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('my_movies.csv')
data
```

```python
data=pd.get_dummies(data)
data
```

```python
items=apriori(data,min_support=0.2,use_colnames=True)
items
```

```python
rules=association_rules(items,metric='lift',min_threshold=0.5)
rules
```

```python
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(data=rules,x='support',y='confidence')
```

# min_support = 0.2

# hyperparameter=confidence

```python
rules_confidence=association_rules(items,metric='confidence',min_threshold=0.5)
rules_confidence
```

```python
plt.scatter(data=rules_confidence,x='support',y='confidence')
```
